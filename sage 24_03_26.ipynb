```python
def cantidad_primos_menores(n):
    count = 0 
    for i in range(1,n+1): 
        if is_prime(i): 
            count += 1 
    return(count)
cantidad_primos_menores(50)
```

```python
def suma_primos_menores(n): 
    sum = 0 
    for i in range(1,n+1):
        if is_prime(i):
            sum = i + sum
    return(sum)
suma_primos_menores(50)
```

### Ejemplo por curiosidad

En $\mathbb{F}_q$, el número de elementos primitivos es $\varphi(q)$ (función phi de Euler), luego la probabilidad de encontrar un elemento primitivo al azar es $\frac{\varphi(q)}{q}$. Se define la función $\varphi$ de Euler como el número de enteros coprimos con $q$, es decir:

$$\varphi(q) = |\{1 \leq k \leq n \,\colon\, \mathrm{mcd}(k,q)=1\}|$$

    3. Encontrar todos los elementos primitivos de $L$. Elegir uno, y escribir los elementos de $L^*=L\setminus \{0\}$ como potencias de ese elemento.

```python
K.<x> = GF(2)[]
f = x^3+x+1
is_irred = True

current_deg = 1

while current_deg <= f.degree() // 2 and is_irred:
    for p in K.polynomials(of_degree=current_deg):
            # print(f"Iterando {p}")
            if f % p == 0: 
                print(f" El polinomio {p} divide a {f}")
                is_irred = False
                break
    current_deg +=1
if is_irred: 
    print(f"Pues parece que {f} es irreducible")
```
