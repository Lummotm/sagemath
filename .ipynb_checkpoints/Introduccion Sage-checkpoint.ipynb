### Listas con srange(m,n,r)
Podemos también hacer listas usando el comando $srange(m,n,r)$. Esto crea una lista de enteros de $m$ a $n-1$ con paso $r$.

```python
a=srange(0,3);a #No se incluye el extremo
b=srange(3);b #Si no se especifica el extremo inicial se empieza en 0
a==b 
A = srange(10,1,-2);A #se puede cambiar el paso y hacerlo negativo, en este caso la lista llega hasta n+1
```

## Funciones

```python
def paridad(n): 
    return n%2==0 #La indentación es obligatoria
paridad(15)
```

```python
paridad(-12)
```

```python
def eslasuma(L):
    n=len(L)
    s=L[n-1]
    S=sum(L[:n-1]) #suma de todos los términos menos el último
    return s==S
print(eslasuma([1,2,3]))
print(eslasuma([1,2,3,4,5]))
print(eslasuma([1,2,3,4,5,15]))
```

2) Definir una función que acepte como argumento un número natural n y devuelva como resultado la cantidad de números primos menores o iguales que n.

```python
def primosmenores(n):
    L=[i for i in srange(1,n+1) if is_prime(i)]
    return L
primosmenores(50)
```

```python
def numeroprimos(n):
    return len(primosmenores(n))
numeroprimos(50)
```

```python
def sumaprimos(n):
    return sum(primosmenores(n))
sumaprimos(50)
```

# Cuerpos finitos
En la siguiente página podéis comprobar la documentación de Sage por si tenéis alguna duda:
https://doc.sagemath.org/html/en/reference/finite_rings/sage/rings/finite_rings/finite_field_constructor.html

```python
#Para definir un cuerpo finito primo:
K = GF(3);K;type(K)
#Podemos sacar una lista con los diferentes elementos de K y podemos operar con ellos
K.list(); #Si ahora queremos operar con los elementos de K podemos hacer:
K(2)*K(2) #Esto hace 2*2 mod 3
K(2)+K(1) #Esto hace 2+1 mod 3
[i^3 for i in K] #Comprobamos que a^i  = i mod 3
```

```python
#Otra manera que tenemos de definir los cuerpos finitos es usando el anillo de enteros modulo un primo
Z3 = Integers(3);type(Z3)
Z3.list()
Z3(2)*Z3(2)
```

```python
#vamos a ver ahora cómo se definen cuerpos finitos cuando q=p^r
L.<a> = GF(9) #La clase de x se llama ahora a, y podemos hacer las operaciones con respecto a eso
L.list()
minpoly(a) #Este es el polinomio mínimo de a, que es lo que Sage ha usado para construir F_9
#Si no le especificamos nada, puede que Sage utilice otro nombre para la clase de x:
L2 = GF(9);L2.list() #Aquí ha llamado z2 a la variable
L==L2 #Además para Sage son cuerpos diferentes
```

```python
L.cardinality();L2.cardinality()
```

### Ejemplo por curiosidad

En $\mathbb{F}_q$, el número de elementos primitivos es $\varphi(q)$ (función phi de Euler), luego la probabilidad de encontrar un elemento primitivo al azar es $\frac{\varphi(q)}{q}$. Se define la función $\varphi$ de Euler como el número de enteros coprimos con $q$, es decir:

$$\varphi(q) = |\{1 \leq k \leq n \,\colon\, \mathrm{mcd}(k,q)=1\}|$$


```python
Euler =[(n,euler_phi(n)/n) for n in srange(1,1000) if n.is_prime_power()];list_plot(Euler)
```

## Seguimos con cuerpos finitos


```python
#Podemos definir los cuerpos finitos usando un polinomio (como en clase)
F.<a> = GF(9, modulus = x^2+2*x+2);F
for j in F:
    j,minpoly(j)
```

```python
#Sage también es capaz de sacar las tablas de suma y multiplicación de los cuerpos finitos
R.<a>=GF(8,modulus='primitive')
R.addition_table(names='elements')
R.multiplication_table(names='elements')
```

## Anillos de polinomios

```python
#Podemos también trabajar con anillos de polinomios y, por supuesto, los coeficientes se oueden tomar en un cuerpo finito.
K.<x>=PolynomialRing(QQ) #Tanto K como F definen el anillo de polinomios con coeficientes en el cuerpo de los racionales
F.<x>=QQ[] 
K==F
f = x^2-1; f.parent() #Nos dice donde está f
#Podemos factorizar polinomios, operar con ellos...
factor(f); f.roots() #Esto da la factorización de f y las raíces de f con sus multiplicidades
g = x^3+2*x-2
#¡Vamos a hacer algunas operaciones!
f*g; gcd(f,g); g//f; g%f
f(2) #Se puede evaluar también
```

```python
#Y todo esto también se puede hacer sobre cuerpos finitos
B.<x>=GF(5)[] #Sobre un cuerpo finito
f=x^2+1; f.parent()
factor(f)

#También sobre cuerpos finitos no primos
F.<a>=GF(9,modulus='primitive')
A.<x>=F[] #Si queremos especificar el nombre del generador y demás, esta es la manera correcta de hacerlo
A
f=x^2+1
factor(f)
```

```python
#También se puede definir todo directamente como cocientes
A.<y>=GF(2)[]
f=y^3+y+1
B.<x>=A.quotient(f) 
[j for j in B]
B.is_field()

B.addition_table(names='elements')
```
