```python
print()
print("=== Ejercicio 3 ===")
print()



def decodificacion_r(r, X, K):
    # Parámetros del código
    t = 2
    n = 15
    k = 10
    
    # FÓRMULAS CORRECTAS (Welch-Berlekamp)
    l_0 = n - 1 - t  # Grado máximo de Q0
    l_1 = n - 1 - t - (k - 1)  # Grado máximo de Q1
    
    # Construcción de la matriz A (para Q0)
    A_elems = [
        vector(K, [X[j]^i for i in srange(l_0 + 1)])
        for j in srange(n) 
    ]
    A = matrix(K, A_elems)
    
    # Construcción de la matriz B (para Q1, multiplicando por r[j])
    B_elems = [
        vector(K, [r[j] * X[j]^i for i in srange(l_1 + 1)])
        for j in srange(n)
    ]
    B = matrix(K, B_elems)
    
    # Formamos la matriz del sistema M = [A | B]
    M = block_matrix([[A, B]])
    
    # Resolvemos el sistema buscando el núcleo
    coef = M.right_kernel()
    if coef.dimension() == 0:
        print("Error: No se encontró solución para el sistema.")
        return
        
    K_vec = coef.basis()[0] # Primera solución no nula
    
    # Separamos los coeficientes
    q0_coefs = K_vec[:l_0 + 1]
    q1_coefs = K_vec[l_0 + 1:]
    
    # Definimos el anillo de polinomios para g(x)
    R_pol.<x_pol> = PolynomialRing(K)
    Q0 = R_pol(list(q0_coefs))
    Q1 = R_pol(list(q1_coefs))
    
    # 4. Localización de errores
    posiciones_error = [j for j in range(n) if Q1(X[j]) == 0]
    print(f"Raíces del localizador Q1 en índices: {posiciones_error}")
    
    # 5. Recuperación del polinomio original g(x) = -Q0 / Q1
    # En GF(2^m) el menos es opcional, pero lo ponemos por rigor
    g = -Q0 // Q1
    print(f"Polinomio decodificado g(x): {g}")
    
    # 6. Comprobación de distancia
    c_decod = vector(K, [g(X[j]) for j in range(n)])
    distancia = sum(1 for j in range(n) if r[j] != c_decod[j])
    
    print(f"Distancia entre r y c_decod: {distancia}")
    if distancia <= t:
        print("Resultado: Decodificación CORRECTA.")
    else:
        print("Resultado: ERROR (Más de 2 errores).")
    print()
```
