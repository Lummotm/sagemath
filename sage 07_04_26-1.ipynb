# Códigos lineales en Sage

Hay distintos métodos para construir un código en Sage:
- A partir de una matriz generadora: C = LinearCode(G)
- A partir de una matriz de control: C = codes.from_parity_check_matrix(H)
- Como un código aleatorio sobre el cuerpo $\mathbb{F}_q$ de longitud $n$ y dimensión $k$: C = codes.random_linear_code(GF(q),n, k)

## Espectro de un código
- Ya hemos visto que C.minimum_distance() devuelve la distancia mínima de $C$.

Este método hace una búsqueda exhaustiva y no es muy eficiente en general.

Sea $\mathcal C$ un código de longitud $n$. El espectro de un código, también llamado distribución de pesos, es la $n$-tupla $(A_0, \ldots, A_n)$ donde $A_w$ es el número de palabras de $\mathcal C$ con peso $w$.

- La función C.spectrum() devuelve el espectro de C.

EL polinomio de pesos homogéneo de $\mathcal C$ se define de la siguiente manera:

$$W_{\mathcal C}(X,Y) = \sum_{w=0}^n A_w X^{w}Y^{n-w}$$

Los polinomios de pesos satisfacen la identidad de McWilliams:

$$W_{\mathcal C^{\perp}}(X,Y) = q^{-k}W_{\mathcal C}(Y-X,Y+(q-1)X)$$

- La función C.weight_enumerator() devuelve el polinomio de pesos de C.
- La función C.support() devuelve los pesos para los cuales existen palabras en $C$ de dicho peso


```python
C=codes.random_linear_code(GF(3),70,15)
%time C.minimum_distance() #Devuelve el tiempo
```

Vamos a hacer una función para descodificar por síndrome que calcule la clase de un vector 
y su líder para un código binario (realmente no nos hace falta porque Sage ya tiene incluida una función que decodifica)

```python
def clase(v,C):
    L=[]
    for i in C.list(): #construimos v+C
        L.append(i+v) 
    return L

def lider(A):
    B=copy(A)
    lid=B[0]
    wd=lid.hamming_weight()
    L=[0 for i in srange(len(lid)+1)]
    for i in B:
        w=i.hamming_weight()
        L[w]=L[w]+1
        if w<wd:
            lid=i
            wd=w
    return [L,lid]
```

## Códigos Reed-Solomon

-Se pueden construir los códigos Reed-Solomon clásicos con la función codes.ReedSolomonCode(base_field, length, dimension, primitive_root=None). Estos códigos son códigos Reed-Solomon, cuya longitud n debe dividir a q-1, y cuyos puntos de evaluación son $\alpha_i=\alpha^{i-1}$, $1\leq i \leq n$, donde $\alpha$ es una raíz primitiva de $\mathbb{F}_q$.

-Si queremos construir códigos Reed-Solomon que no sean clásicos, tenemos que usar la función de los códigos Reed-Solomon generalizados: codes.GeneralizedReedSolomonCode(evaluation_points, dimension, column_multipliers=None). Dados $n$ elementos de $\mathbb{F}_q$, $\alpha_1,\dots,\alpha_n$, y dados $n$ elementos no nulos de $\mathbb{F}_q$, $\beta_1,\dots,\beta_n$, el código Reed-Solomon generalizado de dimensión $k$ y longitud $n$ es el conjunto:
$$
\{ (\beta_1f(\alpha_1),\dots,\beta_nf(\alpha_n))\mid f\in \mathbb{F}_q[x], \deg f< k\}
$$
Luego los Reed-Solomon que hemos visto en clase son los generalizados con $\beta_i=1 \forall i$.

```python
F = GF(11)
n, k = 12, 5
F.list()[:n]
```

```python
C = codes.GeneralizedReedSolomonCode(F.list()[:n], k); C #al no especificar los betas, se entiende que son 1
```

```python
G=C.generator_matrix(); G
# Es MDS: d=n-k+1
C.minimum_distance()
```

```python
D = codes.decoders.GRSBerlekampWelchDecoder(C); D #Es el algoritmo que hemos visto en clase
# Otra manera: D=C.decoder("BerlekampWelch")
# Este código puede corregir hasta 3 errores
t=floor((C.minimum_distance()-1)/2); t

codeword = C.random_element(); codeword

word=codeword+vector(GF(11),[0, 1, 2, 3, 0, 0, 0, 0, 0, 0, 0]); word

D.decode_to_code(word), D.decode_to_code(word)==codeword
```

## Construcciones de códigos a partir de otros
Hay varias construcciones, dos ejemplos son:

   -Código recortado: Se hace la intersección de $C$ con $\{x_i\}=0$, y luego se elimina la coordenada i-ésima. La idea es mantener la redundancia y reducir la longitud y la dimensión. Se usa la función: shortened(L) (recortar en las coordenadas de la lista L)
   
   -Código "punctured": Se elimina directamente la coordenada i-ésima. La idea es manener la dimensión, y reducir la longitud y la redundancia. Se usa la función: punctured(L) (eliminar las coordenadas de la lista L)
   
No siempre se cumple lo que digo como idea, en casos particulares puede que baje la dimensión al hacer el "punctured", y en algunos casos puede que no baje la dimensión al hacer el recortado (por ejemplo si todas las palabras ya tienen una coordenada igual a 0).
